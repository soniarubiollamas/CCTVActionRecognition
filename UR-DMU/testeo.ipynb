crear nuevo test_train.list que no contenga los archivos con diferente tamaño

```python
# Importar las librerías necesarias
import os

# Función para leer un archivo .list y obtener los nombres de los archivos
def leer_archivo_list(ruta_archivo):
    with open(ruta_archivo, 'r') as f:
        lineas = f.readlines()
    # Eliminar cualquier salto de línea y devolver una lista de los archivos
    return [linea.strip() for linea in lineas]

# Función para eliminar los archivos que están en la lista de exclusión (incluyendo variantes)
def filtrar_archivos(original_files, exclusion_files):
    archivos_filtrados = []
    for archivo in original_files:
        # Si ningún archivo de la lista de exclusión es un prefijo del archivo actual, lo mantenemos
        if not any(os.path.basename(archivo).startswith(os.path.splitext(exclusion)[0]) for exclusion in exclusion_files):
            archivos_filtrados.append(archivo)
    return archivos_filtrados

# Función para escribir los archivos filtrados en un nuevo archivo .list
def escribir_archivo_list(ruta_archivo, archivos):
    with open(ruta_archivo, 'w') as f:
        for archivo in archivos:
            f.write(archivo + '\n')

# Rutas de los archivos de entrada y salida
ruta_original_list = 'list/Test_Train_excluido.list'  # Reemplaza con la ruta correcta
ruta_exclusion_list = 'list/excluir_URDMU.list'  # Reemplaza con la ruta correcta
ruta_nuevo_list = 'list/Test_Train_excluido_URDMU.list'  # Reemplaza con la ruta correcta

# Leer los archivos desde los .list
archivos_originales = leer_archivo_list(ruta_original_list)
archivos_a_excluir = leer_archivo_list(ruta_exclusion_list)

# Filtrar los archivos (eliminar los que tienen el mismo prefijo que en la lista de exclusión)
archivos_filtrados = filtrar_archivos(archivos_originales, archivos_a_excluir)

# Escribir el nuevo archivo .list con los archivos filtrados
escribir_archivo_list(ruta_nuevo_list, archivos_filtrados)

# Mostrar resultado en la salida
print(f"Se ha creado un nuevo archivo con {len(archivos_filtrados)} archivos filtrados en '{ruta_nuevo_list}'.")

```

Guardar solo los .npy de test de las nuevas features descargadas

```python
import os
import numpy as np

# Rutas de las carpetas
ruta_features = 'test/15'  
ruta_test = 'test'  
ruta_output = 'test/15'  

# Obtener la lista de nombres de los archivos de test a partir de la carpeta de preds
nombres_test = [f.split('_x264')[0] for f in os.listdir(ruta_test) if f.endswith('.npy')]

# Crear la carpeta de salida si no existe
os.makedirs(ruta_output, exist_ok=True)

# Mantener un conjunto de nombres únicos
nombres_guardados = set()

# Recorrer la carpeta de features
for nombre in nombres_test:
    # Definir la ruta de la carpeta de features correspondiente
    nombre = f'{nombre}_x264'
    carpeta_features = os.path.join(ruta_features, nombre)
    
    # Verificar si la carpeta existe y si contiene el archivo 'features.npy'
    archivo_features = os.path.join(carpeta_features, 'feature.npy')
    if os.path.exists(archivo_features):
        # Cargar el archivo .npy
        features = np.load(archivo_features)
        
        # Guardar las features con el nombre correspondiente en la carpeta de salida
        archivo_salida = os.path.join(ruta_output, f"{nombre}.npy")
        np.save(archivo_salida, features)
        
        # Añadir el nombre al conjunto de nombres únicos
        nombres_guardados.add(archivo_salida)
        print(f"{archivo_salida}")
    else:
        print(f"No se encontró 'features.npy' en la carpeta: {carpeta_features}")

# Guardar los nombres únicos en un archivo de texto
with open('test/rgb_test/nombres_test.txt', 'w') as f:
    for nombre in sorted(nombres_guardados):
        f.write(f"{nombre}\n")

```

Comarar tamaños de los archivos .npy

```python
import os
import numpy as np

# Rutas de las carpetas de los archivos originales y nuevos
ruta_og = 'test'  # Cambia esta ruta por la correspondiente a tus archivos originales
ruta_new = 'test/rgb_test'  # Cambia esta ruta por la correspondiente a tus nuevos archivos

# Obtener la lista de archivos de ambas carpetas
archivos_og = [f for f in os.listdir(ruta_og) if f.endswith('.npy')]
archivos_new = [f for f in os.listdir(ruta_new) if f.endswith('.npy')]

# Lista para almacenar los archivos con tamaños diferentes
archivos_diferentes = []

# Comparar tamaños de cada archivo
for archivo in archivos_og:
    # Verificar si el archivo correspondiente existe en la carpeta nueva
    if archivo in archivos_new:
        ruta_og_archivo = os.path.join(ruta_og, archivo)
        ruta_new_archivo = os.path.join(ruta_new, archivo)
        
        # Cargar ambos archivos .npy
        og = np.load(ruta_og_archivo)
        new = np.load(ruta_new_archivo)
        
        # Comparar las formas (shapes) de los archivos
        if og.shape != new.shape:
            archivos_diferentes.append((archivo, og.shape, new.shape))

# Mostrar los archivos que tienen tamaños diferentes
if archivos_diferentes:
    print("Archivos con tamaños diferentes:")
    for archivo, shape_og, shape_new in archivos_diferentes:
        print(f"{archivo}: original {shape_og}, nuevo {shape_new}")
else:
    print("Todos los archivos tienen el mismo tamaño.")
```

```python
import os

# Ruta al archivo .list
list_file = 'list/UCF_Train.list'

# Ruta a la carpeta principal de 'train'
train_folder = 'UCF_Crime_Frames/train'

# Leer la lista de archivos desde el .list
with open(list_file, 'r') as f:
    file_list = f.read().splitlines()

# Archivos esperados para el set de entrenamiento
expected_files = set()

# Procesar cada línea en la lista
for file_path in file_list:
    # Nos interesa solo los archivos que empiezan con "train"
    if file_path.startswith('train/'):
        # Extraer el nombre del archivo sin ruta y sin extensión .npy
        file_name = os.path.basename(file_path).replace('.npy', '')
        if file_name in expected_files:
            continue
        else:
            expected_files.add(file_name)
    
# Verificar archivos faltantes


missing_files = []

# Recorrer las carpetas dentro de "train"
for root, dirs, files in os.walk(train_folder):
    for file in files:
        # Remover la extensión del archivo para compararlo
        file_name = os.path.splitext(file)[0]
        if file_name not in expected_files:
            print(f'Archivo faltante: {file_name}')
            missing_files.append(file_name)

# Resumen de archivos faltantes
if missing_files:
    print(f'Se encontraron {len(missing_files)} archivos faltantes en la carpeta train.')
else:
    print('Todos los archivos de la lista están presentes en la carpeta train.')
```

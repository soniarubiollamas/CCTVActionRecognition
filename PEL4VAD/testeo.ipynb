```python
import os

def obtener_nombres_archivos_carpeta(ruta_carpeta):
    """Obtiene una lista con los nombres de los archivos de una carpeta"""
    return set(os.listdir(ruta_carpeta))

def obtener_nombres_desde_txt(ruta_txt):
    """Obtiene los nombres de los archivos desde un archivo txt"""
    nombres_txt = set()
    with open(ruta_txt, 'r') as archivo:
        for linea in archivo:
            nombre = linea.split()[0]  # El nombre del archivo est치 en la primera columna
            nombres_txt.add(nombre)
    return nombres_txt

def nombres_no_en_carpeta(ruta_carpeta, ruta_txt):
    """Devuelve los nombres que est치n en el archivo txt pero no en la carpeta"""
    nombres_archivos = obtener_nombres_archivos_carpeta(ruta_carpeta)
    nombres_txt = obtener_nombres_desde_txt(ruta_txt)
    
    # Devuelve los nombres que est치n en el archivo txt pero no en la carpeta
    return nombres_txt - nombres_archivos
```

```python
import os
from moviepy.editor import VideoFileClip

def obtener_duracion_total_videos(ruta_carpeta):
    """Calcula el tiempo total de duraci칩n de todos los videos en la carpeta"""
    duracion_total = 0
    # Recorre todos los archivos en la carpeta
    for archivo in os.listdir(ruta_carpeta):
        # Si el archivo es un video (puedes a침adir m치s extensiones si es necesario)
        if archivo.endswith(('.mp4', '.avi', '.mkv', '.mov', '.flv')):
            ruta_video = os.path.join(ruta_carpeta, archivo)
            try:
                # Cargar el video y obtener su duraci칩n
                video_clip = VideoFileClip(ruta_video)
                duracion_total += video_clip.duration  # Duraci칩n en segundos
                video_clip.close()
            except Exception as e:
                print(f"No se pudo procesar el archivo {archivo}: {e}")
    
    return duracion_total

# Define la ruta de la carpeta
ruta_carpeta = 'E:/XONI_MASTER/1_interships/PEL4VAD/videos'  # Ejemplo: '/home/usuario/videos'

# Obtener la duraci칩n total en segundos
duracion_total = obtener_duracion_total_videos(ruta_carpeta)

# Convertir la duraci칩n total a horas, minutos y segundos
horas = int(duracion_total // 3600)
minutos = int((duracion_total % 3600) // 60)
segundos = int(duracion_total % 60)

print(f"Duraci칩n total de los videos: {horas} horas, {minutos} minutos, {segundos} segundos")
```

```python
import numpy as np
import os

mode = "test"
mode = "train"
# Ruta de la carpeta donde est치n las predicciones para cada algoritmo
carpetas_algoritmos = {
    'PEL4VAD': f'C:/Users/xoni/Documents/1_uni/1_insterships/cuda_version/PEL4VAD-master/predictions/{mode}',
    'URDMU': f'C:/Users/xoni/Documents/1_uni/1_insterships/UR-DMU/frame_label/predictions/{mode}U',
    'BNWVAD': f'C:/Users/xoni/Documents/1_uni/1_insterships/BN-WVAD-main/frame_label/predictions/{mode}'
}

gt_path = f'C:/Users/xoni/Documents/1_uni/1_insterships/cuda_version/PEL4VAD-master/predictions/gt/'  # Ruta al archivo de ground truth

# Funci칩n para cargar predicciones y ground truth de cada video
def cargar_predicciones_y_gt(carpeta_algoritmos, carpeta_gt):
    predicciones = {}
    gt = {}

    # Iterar sobre las carpetas de predicciones de cada algoritmo
    for nombre_algoritmo, carpeta in carpeta_algoritmos.items():
        archivos_predicciones = [f for f in os.listdir(carpeta) if f.endswith('.npy')]
        
        for archivo in archivos_predicciones:
            ruta_completa = os.path.join(carpeta, archivo)
            nombre_video = archivo.split('_pred.npy')[0]
            
            # Cargar las predicciones en la estructura de diccionarios
            if nombre_video not in predicciones:
                predicciones[nombre_video] = {}
            predicciones[nombre_video][nombre_algoritmo] = np.load(ruta_completa)
            
            # Cargar el ground truth correspondiente al video
            if nombre_video not in gt:
                ruta_gt = os.path.join(carpeta_gt, f"{nombre_video}_gt.npy")
                if os.path.exists(ruta_gt):
                    gt[nombre_video] = np.load(ruta_gt)
                else:
                    print(f"Ground truth para {nombre_video} no encontrado.")

    return predicciones, gt

# Funci칩n de votaci칩n de consenso (Opci칩n 1)
def votacion_anomalias_multiples_modelos(predicciones, gt, threshold=0.5):
    for video, preds in predicciones.items():
        if video not in gt:
            print(f"Ground truth no disponible para el video {video}.")
            continue
        
        consenso_total = np.zeros(len(gt[video]))
        
        for algoritmo, prediccion in preds.items():
            prediccion_binaria = (prediccion >= threshold).astype(int)
            consenso_total += prediccion_binaria  # Sumar el consenso

        # Contar las coincidencias
        total_frames = len(gt[video])
        coincidencias_0 = np.sum(consenso_total == 0)
        coincidencias_1 = np.sum(consenso_total == 1)
        coincidencias_2 = np.sum(consenso_total == 2)
        coincidencias_3 = np.sum(consenso_total == 3)

        print(f"Video: {video}")
        print(f"Total de frames: {total_frames}")
        print(f"Ning칰n algoritmo detecta anomal칤a: {coincidencias_0} frames")
        print(f"Un algoritmo detecta anomal칤a: {coincidencias_1} frames")
        print(f"Dos algoritmos detectan anomal칤a: {coincidencias_2} frames")
        print(f"Tres algoritmos detectan anomal칤a: {coincidencias_3} frames")


# Cargar las predicciones y ground truth
predicciones, gt = cargar_predicciones_y_gt(carpetas_algoritmos, gt_path)

# Aplicar votaci칩n de consenso a los modelos
votacion_anomalias_multiples_modelos(predicciones, gt, threshold=0.5)
```

```python
import os
import numpy as np

mode = "test"
mode_BN = "test_normalized"

carpetas_algoritmos = {
    'PEL4VAD': f'C:/Users/xoni/Documents/1_uni/1_insterships/cuda_version/PEL4VAD-master/predictions/{mode}',
    'URDMU': f'C:/Users/xoni/Documents/1_uni/1_insterships/UR-DMU/frame_label/predictions/{mode}',
    'BNWVAD': f'C:/Users/xoni/Documents/1_uni/1_insterships/BN-WVAD-main/frame_label/predictions/{mode_BN}'
}

gt_path = f'C:/Users/xoni/Documents/1_uni/1_insterships/cuda_version/PEL4VAD-master/predictions/gt/test'  # Ruta al archivo de ground truth

# Funci칩n para cargar predicciones y ground truth de cada video
def load_and_concatenate_files(pred_folder, gt_folder, mode="test"):
    preds_list = []
    gt_list = []

    for pred_file in os.listdir(gt_folder):
        pred_file_pred = pred_file.replace("x264_", "")
        name = pred_file_pred.split("_pred")[0]

        ###### si no quiero hacer el infer con los videos normales ######
        if name == "Normal" and mode == "train":
            continue
        ###############################################################

        pred_path = os.path.join(pred_folder, pred_file_pred)
        gt_path = os.path.join(gt_folder, pred_file)

        if os.path.isfile(pred_path) and os.path.isfile(gt_path):
            preds = np.load(pred_path)
            gt = np.load(gt_path)

            if len(preds) < len(gt):
                # truncar gt
                gt = gt[:len(preds)]

            if len(preds) > len(gt):
                gt = np.append(gt, np.repeat(gt[-1], len(preds) - len(gt)))

            preds_list.append(preds)
            gt_list.append(gt)
        else:
            print(f"Pred file or GT file not found for {pred_file}")

    concatenated_preds = np.concatenate(preds_list)
    concatenated_gt = np.concatenate(gt_list)

    return concatenated_preds, concatenated_gt

predictions_PEL4VAD, ground_truth_PEL4VAD = load_and_concatenate_files(carpetas_algoritmos['PEL4VAD'], gt_path)

number_of_anomalies = np.sum(ground_truth_PEL4VAD)
number_of_normal_frames = len(ground_truth_PEL4VAD) - number_of_anomalies
print(f'number of anomalies: {number_of_anomalies}')
print(f'number of normal frames: {number_of_normal_frames}')
print(f'percentage of anomalies: {number_of_anomalies / len(ground_truth_PEL4VAD) * 100}%')
print(f'percentage of normal frames: {number_of_normal_frames / len(ground_truth_PEL4VAD) * 100}%')
```

```python
import os

# Define las rutas de las tres carpetas
mode = "train"
mode_BN = "train_normalized"

folders = {
    'PEL4VAD': f'C:/Users/xoni/Documents/1_uni/1_insterships/cuda_version/PEL4VAD-master/predictions/{mode}',
    'URDMU': f'C:/Users/xoni/Documents/1_uni/1_insterships/UR-DMU/frame_label/predictions/{mode}',
    'BNWVAD': f'C:/Users/xoni/Documents/1_uni/1_insterships/BN-WVAD-main/frame_label/predictions/{mode_BN}'  # Aseg칰rate de ajustar la ruta
}

def eliminar_archivos_no_comunes(folders):
    # Obtener la lista de archivos (sin extensiones) para cada carpeta
    archivos_por_carpeta = {
        folder_name: {os.path.splitext(f)[0] for f in os.listdir(folder_path) if f.endswith('.npy')}
        for folder_name, folder_path in folders.items()
    }

    # Obtener la intersecci칩n de todos los conjuntos para encontrar los archivos comunes
    archivos_comunes = set.intersection(*archivos_por_carpeta.values())

    # Eliminar los archivos que no est치n en la intersecci칩n
    for folder_name, folder_path in folders.items():
        for file_name in os.listdir(folder_path):
            file_base_name = os.path.splitext(file_name)[0]
            if file_base_name not in archivos_comunes and file_name.endswith('.npy'):
                # Eliminar el archivo si no est치 en los archivos comunes
                file_path = os.path.join(folder_path, file_name)
                os.remove(file_path)
                print(f"Archivo eliminado: {file_path} de la carpeta {folder_name}")

# Ejecutar la funci칩n
eliminar_archivos_no_comunes(folders)
```

```python
import numpy as np
import os

# Ruta donde est치n almacenados los archivos de predicciones de cada algoritmo
# Cambia estas rutas seg칰n la ubicaci칩n de tus archivos
algorithms_paths = [
    "C:/Users/xoni/Documents/1_uni/1_insterships/cuda_version/PEL4VAD-master/predictions/test", 
    "C:/Users/xoni/Documents/1_uni/1_insterships/UR-DMU/frame_label/predictions/test", 
    "C:/Users/xoni/Documents/1_uni/1_insterships/BN-WVAD-main/frame_label/predictions/test_normalized"
]

# Ruta donde se guardar치n las predicciones combinadas
output_path = "predictions/test/combined_3_algorithms"

# Aseg칰rate de que el directorio de salida existe
os.makedirs(output_path, exist_ok=True)


video_files = sorted([f for f in os.listdir(algorithms_paths[0]) if f.endswith('.npy')])


# Procesar cada video
for video_file in video_files:
    # Cargar las predicciones de cada algoritmo
    preds = []
    for path in algorithms_paths:
        pred = np.load(os.path.join(path, video_file))  # Cargar archivo de cada algoritmo
        preds.append(pred)
    
    # Asegurarse de que todas las predicciones tienen la misma dimensi칩n X
    X = preds[0].shape[0]
    # check if all preds have the same shape
    if all(pred.shape[0] != X for pred in preds):
        raise ValueError(f"Las dimensiones no coinciden para el archivo {video_file}")
    
    # Crear el array de salida con formato [3, X, 1024]
    combined_pred = np.stack(preds, axis=0)

    # Guardar el archivo combinado
    np.save(os.path.join(output_path, video_file), combined_pred)

print("Predicciones combinadas guardadas en:", output_path)
```
